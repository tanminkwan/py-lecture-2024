```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#  
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Pandas DataFrame 
df = pd.DataFrame({'X': x, 'Y': y})

# 
plt.figure(figsize=(10, 6))
plt.plot(df['X'], df['Y'], label='sin(x)')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sin function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#   
np.random.seed(42)
data = {
    'A': np.random.randn(100),
    'B': np.random.rand(100) * 100,
    'C': np.random.choice(['X', 'Y', 'Z'], 100)
}
df = pd.DataFrame(data)

#   
grouped_df = df.groupby('C').mean()

# 
plt.figure(figsize=(10, 6))
grouped_df['B'].plot(kind='bar', color='skyblue')
plt.xlabel('Category')
plt.ylabel('Mean Value of B')
plt.title('Mean Value of B by Category')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#   
dates = pd.date_range('2022-01-01', periods=50)
values = np.random.randn(50)

#   DataFrame 
df = pd.DataFrame({'Date': dates, 'Value': values})

#   
df.set_index('Date', inplace=True)

#  
df['MA'] = df['Value'].rolling(window=5).mean()

# 
plt.figure(figsize=(10, 6))
plt.plot(df['Value'], label='Original Data')
plt.plot(df['MA'], label='Moving Average (window=5)')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Timely Moving Average')
plt.legend()
plt.grid(True)
plt.show()
```
